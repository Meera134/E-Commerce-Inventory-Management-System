What is Inventory Management?\
Inventory management refers to tracking the goods a company has in stock. It includes such activities as ordering, restocking, storing and inventory forecasting.

* **An e-commerce inventory management,would detail the process of tracking, storing, and managing the stock of products sold online by an e-commerce business, including crucial aspects like demand forecasting, stock levels optimization, order fulfillment, and real-time visibility to ensure efficient operations and customer satisfaction, often utilizing specialized software to automate these functions across multiple sales channels.**

The E-Commerce Product Inventory Management System is designed to efficiently manage an online product catalog using a Binary Search Tree (BST). This system enables quick search, insertion, and deletion of products, while also supporting,sorting and filtering based on various attributes such as price, category, availability, and more.

<b>BINARY SEARCH TREE </b>
* A Binary Search Tree is a type of data structure that organises elements in such a way that it allows operations like insertion, deletion,traversal(inorder traversal) and searching with the data having no duplicate values.\
  3 classes are used
  <ul>
      <li>class Product</li>
      <li>class BSTNode</li>
      <li>class InventoryManagement</li>
  </ul>
* <b> classProduct</b> -The `Productclass defines an item in the inventory with an ID, name, price, category, and stock, allows creating new products, and provides a method to display product details clearly.\
-Attributes are: product id name price category stock in this we are using Constructor which initializes attributes that were mentioned.
* <b> classBSTNode</b>-The BSTNode class stores a Product and connects to left and right nodes, organizing products in a Binary Search Tree (BST) for quick search, insert, and delete.\
-The key reference for insertion in the BST is the Product ID (id), which ensures unique identification and maintains the BST order for efficient searching, insertion, and deletion.<b>left:</b>points to smaller value node <b>right:</b>points to larger Value node.
* <b> The InventoryManagement class</b>-serves as the main entry point for managing the e-commerce product inventory using a Binary Search Tree (BST).\

 ## OPERATIONS ##
1.<b> Insertion (Adding a New Product)</b>\
Adds a new product based on its Product ID.\
Smaller ID → Left subtree, Larger ID → Right subtree.\
Time Complexity:$O(logn)$.

2.<b> Search (Finding a Product by ID)</b>\
Quickly locates a product in the inventory.\
Compare ID → Move left or right → Find product.\
Time Complexity: $O(logn)$.


3.<b> Deletion (Removing a Product from Inventory)</b>\
Removes a product when out of stock or discontinued.\
Three Cases:\
No child → Remove node.\
One child → Replace with child.\
Two children → Replace with inorder successor.\
Time Complexity: $O(logn)$.


4.<b> Sorting (Displaying Products in Order)</b>\
Uses in-order traversal (Left → Root → Right) for sorting.\
Time Complexity:$O(n)$.


5.<b>Filtering (Sorting by Category, Price, Stock)</b>\
Retrieves products based on specific criteria.\
BST traversal with conditions (e.g., price range, stock level).\
Time Complexity:$O(n)$.
